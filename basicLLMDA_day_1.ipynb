###Sentiment Analysis using Transformers Models


````python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much"
]
classifier(inputs)
````

This cell installs the `transformers` library and its `sentencepiece` dependencies.

### Text Generation Model (like chatgpt)


````python
generator = pipeline("text-generation")
generator("Once upon a time in National College of Engineering, We are Learning LLM")
````

````python
qa = pipeline("question-answering")
qa({
    "question": "Where do I learn LLM?",
    "context": "Once upon a time in National College of Engineering, We are Learning LLM....",
})
````

### Language Translator


````python
translator = pipeline("translation_en_to_de")
translator("I love programming Language.")
````

#### NER Models


````python
ner = pipeline("ner",model="dbmdz/bert-large-cased-finetuned-conll03-english" )
ner(["My name is Aayush Kandel and I currently study at National College of Engineering"])
````
